```python
def is_valid(row):
    # Get the last code after the last '#'
    if pd.isna(row['HIERARCHY']) or pd.isna(row['CODE']):
        return False
    last_code = row['HIERARCHY'].split('#')[-1]
    code = str(row['CODE'])
    # Check if CODE starts with last_code (2 or 3 chars) and is followed by numbers
    if code.startswith(last_code):
        suffix = code[len(last_code):]
        return suffix.isdigit()
    # Try with only first 2 chars of last_code
    if code.startswith(last_code[:2]):
        suffix = code[2:]
        return suffix.isdigit()
    return False
```

```python
import pandas as pd

df = pd.read_csv("../data/results/duplicates_filtered.csv")
# Apply the cleaning steps
df['Valid'] = df['Valid'].astype(str).str.strip().str.lower()

# Map the known boolean strings. Any other string will result in NaN.
df['Valid'] = df['Valid'].map({
    'true': True,
    'false': False
})

# Now, explicitly fill any NaN values (which originated from unmapped strings)
# and then cast to bool.
# For example, let's assume any unmapped value should be considered False.
df['Valid'] = df['Valid'].fillna(False).astype(bool)

print("\nFinal 'Valid' column content (after cleaning, mapping, filling, and final cast):")
print(df['Valid'])
print("Final 'Valid' column dtype:", df['Valid'].dtype)
```

```python
pairs = []

for i in range(0, len(df), 2):
    pair = df.iloc[i:i+2]

    if not pair.iloc[0]['Valid'] and not pair.iloc[1]['Valid']:
        continue
    
    if pair.iloc[0]['Valid']:
        old = pair.iloc[1]['CODE']
        corrected = pair.iloc[0]['CODE']
    else:
        old = pair.iloc[0]['CODE']
        corrected = pair.iloc[1]['CODE']
        
    pairs.append({'Old': old, 'Corrected': corrected})

result = pd.DataFrame(pairs)
result.to_csv("../data/results/corrected_filtering.csv", index=False)
```
